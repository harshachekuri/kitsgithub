### patterns
### start ^
### [a-z,0-9,._]set os chr
### {9}
### end----> $

```python
s="Goodmorning"
pt="^[A-Za-z]{3,11}$"
# given is 11 word string 
# 3,11 means len is =>3 or <=11
if re.match(pt,s):
    print("True")
else:
    print("False")
```

```python
#a="^[0-9]{10}$"(for any 10 digit )
#b="9490965774"# gives true
# to start with 6,7,8,9
a="^[6-9]{1}[0-9]{9}$"
b="7670819765"# if starting no is 5 false
if re.match(a,b):
    print("true")
else:
    print("false")
```

```python
# phone no validation
a="^[6-9]{1}[0-9]{9}|^[+][9][1][6-9]{1}[0-9]{9}$"
b="+917670819765"# if starting no is 5 false
if re.match(a,b):
    print("true")
else:
    print("false")
```

```python
a="^[6-9]{1}[0-9]{9}|^[+][9][1]$"
# "^" is used
b="917670819765"# if starting no is 5 false
if re.match(a,b):
    print("true")
else:
    print("false")
```

```python
# email
a="^[a-zA-Z_.0-9]{4,42}[@][a-z]{3,8}[.][a-z]{2,9}$"
b="joeldasari10@gmail.com"
if re.match(a,b):
    print ("ok")
else :
    print ("no")
```
